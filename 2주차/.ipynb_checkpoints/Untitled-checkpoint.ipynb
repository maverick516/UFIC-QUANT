```python
# 1. A컬럼의 0번 index에 위치한 값을 선택하시오
print(df["A"].loc[0])

print(df["A"].iloc[0])

# 2. A컬럼의 0~10번 index에 속한 값들을 선택하시오
print(df["A"].loc[0:10])
print(df["A"].iloc[0:11])

# 3. A컬럼부터 C컬럼까지만 선택하시오
print(df.loc[:, "A":"C"])
print(df.iloc[:, 0:3])
```

```python
import random
random.seed(100)
num_df = df.apply(lambda x: df.apply(lambda x: random.randint(0, 101), axis=1), axis=0)
num_df
```

```python
# num_df의 A 와 B값을 합친 D 컬럼을 만드시오.

print(num_df)
num_df["D"] = num_df["A"] + num_df["B"]
print(num_df)

# num_df의 A~D 까지의 sum인 sum 컬럼을 만드시오
num_df["E"] = num_df.sum(axis=1)
print(num_df)

# 각 컬럼 값들의 상관계수를 구하시오
covAB = sum((num_df["A"] - num_df["A"].mean()) * (num_df["B"] - num_df["B"].mean())) / len(num_df)
print(covAB)

standard_D_A = (sum((num_df["A"] - num_df["A"].mean()) ** 2) / len(num_df)) ** (1 / 2)

standard_D_B = (sum((num_df["B"] - num_df["B"].mean()) ** 2) / len(num_df)) ** (1 / 2)

corrAB = covAB / (standard_D_A * standard_D_B)
print(corrAB)

print(num_df.corr())
```

```python
# apply 기능
# why?
# 기본적으로 컴퓨팅 시간을 가장 많이 잡아먹는 것은 주소로의 참조
# for loop 경우 주소로의 이동이 빈번하다
# apply 경우 object 접근이기에 주소로의 이동이 적다.

from time import time
df = pd.DataFrame(index = range(0, 77), columns = ["A","B","C","D"])

pos1 = time()
for i in range(0, 100):
    for col in df.columns:
        for ind in df.index:
            df[col].loc[ind] = str(ind) + col.lower()
    
print(time()-pos1)

pos1 = time()
for i in range(0,100):
    df.apply(lambda x: df.apply(lambda y: str(y.name), axis = 1) + x.name.lower())
    
print(time()-pos1)
```

```python
code = '035420'  # NAVER
url = 'https://finance.naver.com/item/sise_day.nhn?code=' + code + '&page=1'
agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'
response = requests.get(url = url, headers = {'User-Agent': agent})
_soap = BeautifulSoup(response.text, 'lxml')
response.status_code #200이 호출되면 정상적인 url 사이트
```

```python
import traceback
import pandas as pd

#함수로 만들어본다다
def parse_page(code, page):
    try:
        url = 'https://finance.naver.com/item/sise_day.nhn?code={code}&page={page}'.format(code = code, page = page)
        agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'
        response = requests.get(url = url, headers = {'User-Agent': agent})
        _soap = BeautifulSoup(response.text, 'lxml')
        _df = pd.read_html(str(_soap.find("table")), header = 0)[0]
        _df = _df.dropna()
        return _df
    except:
        print('Error!')
    return None
```

```python
data = parse_page(code, 1)
data = data.sort_values('날짜')
data
```

```python
data.set_index(data['날짜'], inplace=True, drop=True)
data.drop(['날짜'], inplace=True, axis=1)
data
```

```python
import matplotlib.pyplot as plt

# 그래프 결과
plt.plot(data['종가'])
plt.title('Naver Price')
plt.xlabel('Date')
plt.xticks(rotation = 45)
plt.ylabel('Price')
```

```python
# 개별 이미지에 그리는 법:
Title = ['Close', 'Percent Change', 'Open', 'High', 'Low']
print(data.columns[:-1])
print(enumerate(data.columns[:-1]))
for title, i in enumerate(data.columns[:-1]):
    print(title)
    print(i)
    plt.plot(data[i])
    plt.title(Title[title])
    plt.xlabel('Date')
    plt.xticks(rotation=45)
    plt.show()
```
