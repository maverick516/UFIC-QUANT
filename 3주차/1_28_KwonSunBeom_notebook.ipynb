```python
# 티커 리스트 정의
tickers = ['GC=F', 'SI=F', 'TLT', 'XLB', 'XLE', 'XLF', 'XLI', 'XLK', 'XLP', 'XLU', 'XLV', 'XLY', 'SPY']
temp_refs = ['Gold Feb 23', 'Silver Mar 23', '채권', '소재', '에너지', '금융', '산업재', 'IT', '필수소비재', '유틸리티', '헬스케어', '임의소비재', 'SPDR S&P 500 ETF Trust']

refs = {}
for i in range(len(tickers)):
    refs[tickers[i - 1]] = temp_refs[i - 1]


num = len(tickers)

# etf 객체 정의
etf = yf.Tickers(tickers)

# ETF 시계열 가격 데이터 (1달 단위)
data = etf.history(start='2000-01-01', actions=False)['Close']
data = data.resample('M').last()

# 매달 수익률 데이터 (1 기준)
rets = data.pct_change().fillna(0) + 1

print(refs)
# print(rets['^GSPC'])
```

```python
# GSPC 대비 수익률 데이터
rets_rel = rets.copy();

for ticker in tickers :
    for date in rets['SPY'].index :
        rets_rel[ticker][date] = rets[ticker][date] / rets['SPY'][date]

rets
```

```python
# GSPC 대비 수익률 데이터 시각화 (결과적 무의미)
for ticker in tickers :
    print(refs[ticker])
    rets_rel[ticker].plot(figsize=(15, 6))
    plt.show()
```

```python
# GSPC 대비 수익률 데이터 누적 후 시각화
rets_rel_cum = rets_rel.copy()
rets_rel_cum = rets_rel_cum.cumprod()

rets_rel_cum['GC=F'].plot(figsize=(15, 6), color='gold')
rets_rel_cum['SI=F'].plot(figsize=(15, 6), color='silver')
rets_rel_cum['TLT'].plot(figsize=(15, 6), color='black')
rets_rel_cum['SPY'].plot(figsize=(15, 6), color='white')
rets_rel_cum['XLB'].plot(figsize=(15, 6), color='saddlebrown')
rets_rel_cum['XLE'].plot(figsize=(15, 6), color='aqua')
rets_rel_cum['XLF'].plot(figsize=(15, 6), color='orange')
rets_rel_cum['XLI'].plot(figsize=(15, 6), color='red')
rets_rel_cum['XLK'].plot(figsize=(15, 6), color='violet')
rets_rel_cum['XLP'].plot(figsize=(15, 6), color='purple')
rets_rel_cum['XLU'].plot(figsize=(15, 6), color='lime')
rets_rel_cum['XLV'].plot(figsize=(15, 6), color='blue')
rets_rel_cum['XLY'].plot(figsize=(15, 6), color='green')
plt.legend()
plt.show()
```

```python
# 모멘텀 전략
rets_dual = pd.DataFrame(index = rets['SPY'].index, columns = ["DUAL"])

for year in range(2001, 2023) :
    for month in range(1, 13) :
        prev_year = year - 1
        if(rets['SPY'].loc[f'{prev_year}-{month}':f'{year}-{month}'].cumprod().tail(1)[0] > rets['TLT'].loc[f'{prev_year}-{month}':f'{year}-{month}'].cumprod().tail(1)[0]) :
            rets_dual['DUAL'].loc[f'{year}-{month}'] = rets['SPY'].loc[f'{year}-{month}']
        else :
            rets_dual['DUAL'].loc[f'{year}-{month}'] = rets['TLT'].loc[f'{year}-{month}']

rets_dual_cum = rets_dual.dropna(axis=0).cumprod()
rets_dual_cum['DUAL'].plot(figsize=(15, 6))
rets['SPY'].cumprod().plot(figsize=(15, 6))
rets['TLT'].cumprod().plot(figsize=(15, 6))
plt.legend()
plt.show()

# rets_dual.dropna(axis=0)
# print(rets['^GSPC'])
```

```python
# 모멘텀 전략 + 할로윈 전략
rets_fin = pd.DataFrame(index = rets['SPY'].index, columns = ['FIN'])
rets_fin['FIN'] = rets_dual['DUAL'].copy()

for year in range(2000, 2023) :
    for month in range(5, 11) :
        rets_fin['FIN'].loc[f'{year}-{month}'] = 1

rets_fin_cum = rets_fin.dropna(axis=0).cumprod()

rets_dual_cum['DUAL'].plot(figsize=(15, 6))
rets_fin_cum['FIN'].plot(figsize=(15, 6))
plt.legend()
plt.show()
```

```python
# 할로윈 전략에 관하여
rets_temp_1 = rets.copy()
rets_temp_2 = rets.copy()
for ticker in tickers :
    for year in range(2000, 2023) :
        for month in range(1, 13) :
            if month in range(5, 11) :
                rets_temp_1[ticker].loc[f'{year}-{month}'] = 1
            else :
                rets_temp_2[ticker].loc[f'{year}-{month}'] = 1
                
rets_temp_1.cumprod().plot(figsize=(15, 6))
rets_temp_2.cumprod().plot(figsize=(15, 6))
plt.legend()
plt.show()
```

```python
# 모멘텀 전략 + 할로윈 전략
rets_fin = pd.DataFrame(index = rets['SPY'].index, columns = ['FIN'])
rets_fin['FIN'] = rets_dual['DUAL'].copy()

for year in range(2000, 2023) :
    for month in range(5, 11) :
        rets_fin['FIN'].loc[f'{year}-{month}'] = rets['TLT'].loc[f'{year}-{month}']

rets_fin_cum = rets_fin.dropna(axis=0).cumprod()

rets_dual_cum['DUAL'].plot(figsize=(15, 6))
rets_fin_cum['FIN'].plot(figsize=(15, 6))
plt.legend()
plt.show()
```

```python
# 모멘텀 전략
rets_dual_new = pd.DataFrame(index = rets['SPY'].index, columns = ["DUAL"])

for year in range(2001, 2023) :
    for month in range(1, 13) :
        prev_year = year - 1
        if(rets['SPY'].loc[f'{prev_year}-{month}':f'{year}-{month}'].cumprod().tail(1)[0] < 1.0 and rets['TLT'].loc[f'{prev_year}-{month}':f'{year}-{month}'].cumprod().tail(1)[0] < 1.0) :
            rets_dual_new['DUAL'].loc[f'{year}-{month}'] = 1
        elif(rets['SPY'].loc[f'{prev_year}-{month}':f'{year}-{month}'].cumprod().tail(1)[0] > rets['TLT'].loc[f'{prev_year}-{month}':f'{year}-{month}'].cumprod().tail(1)[0]) :
            rets_dual_new['DUAL'].loc[f'{year}-{month}'] = rets['SPY'].loc[f'{year}-{month}']
        else :
            rets_dual_new['DUAL'].loc[f'{year}-{month}'] = rets['TLT'].loc[f'{year}-{month}']

rets_dual_new_cum = rets_dual_new.dropna(axis=0).cumprod()
rets_dual_new_cum['DUAL'].plot(figsize=(15, 6))
rets['SPY'].cumprod().plot(figsize=(15, 6))
rets['TLT'].cumprod().plot(figsize=(15, 6))
plt.legend()
plt.show()
```

```python
# 모멘텀 전략 + 할로윈 전략
rets_fin = pd.DataFrame(index = rets['SPY'].index, columns = ['FIN'])
rets_fin['FIN'] = rets_dual_new['DUAL'].copy()

for year in range(2000, 2023) :
    for month in range(5, 11) :
        rets_fin['FIN'].loc[f'{year}-{month}'] = rets['TLT'].loc[f'{year}-{month}']

rets_fin_cum = rets_fin.dropna(axis=0).cumprod()

rets_dual_new_cum['DUAL'].plot(figsize=(15, 6))
rets_dual_cum['DUAL'].plot(figsize=(15, 6))
rets_fin_cum['FIN'].plot(figsize=(15, 6))
plt.legend()
plt.show()
```
